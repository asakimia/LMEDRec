# amazon-book数据预处理

```python
# 读数据
dataset_name = 'ml-100k' # steam,lfm1b-tracks,ml-1m,Amazon_Books

dir_path = '../../'+dataset_name+'/'
inter_path = dir_path+dataset_name+'.inter'
kg_path = dir_path+dataset_name+'.kg'
link_path = dir_path+dataset_name+'.link'

ratings_final = dir_path+dataset_name+'/ratings_final.txt'
kg_final = dir_path+dataset_name+'/kg_final.txt'
item_index2entity_id = dir_path+dataset_name+'/item_index2entity_id.txt'
```

```python
data = pd.read_csv(kg_path, sep="\t")
data_new = data
data_new
```

```python
e_index = 0
r_index = 0
e_map = {}
r_map = {}
```

```python
with open(kg_final, 'w') as fp:
    for index, row in data_new.iterrows():
        h = row['head_id:token']
        t = row['tail_id:token']
        r = row['relation_id:token']
        if not e_map.__contains__(h):
            e_map[h] = e_index
            e_index += 1

        if not e_map.__contains__(t):
            e_map[t] = e_index
            e_index += 1

        if not r_map.__contains__(r):
            r_map[r] = r_index
            r_index += 1
        
        fp.write(str(e_map[h]) + '	' + str(r_map[r]) + '	' + str(e_map[t]))
        fp.write('\r')
```

```python
e_map,r_map
```

```python
data = pd.read_csv(inter_path, sep="\t")
data_new = data
data_new
```

```python
u_index = 0
i_index = 0
u_map = {}
i_map = {}
```

```python
with open(ratings_final, 'w') as fp:
    for index, row in data_new.iterrows():
        u = row['user_id:token']
        i = row['item_id:token']
        r = 1 if row['rating:float']>=3 else 0

        if not u_map.__contains__(u):
            u_map[u] = u_index
            u_index += 1

        if not i_map.__contains__(i):
            i_map[i] = i_index
            i_index += 1
        
        fp.write(str(u_map[u]) + '	' + str(i_map[i]) + '	' + str(r))
        fp.write('\r')
```

```python
u_map,i_map
```

```python
data = pd.read_csv(link_path, sep="\t")
data_new = data
data_new
```

```python
with open(item_index2entity_id, 'w') as fp:
    for index, row in data_new.iterrows():
        i = row['item_id:token']
        e = row['entity_id:token']

        if not i_map.__contains__(i):
            i_map[i] = i_index
            i_index += 1
            print('iiiii'+str(i))

        if not e_map.__contains__(e):
            e_map[i] = e_index
            e_index += 1
            print('eeeee'+str(e))
        
        fp.write(str(i_map[i]) + '	' + str(e_map[e]))
        fp.write('\r')
```
