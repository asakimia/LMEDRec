```python
def build_item_other(row):
    df1 = inter_data[inter_data['item_id:token'] == row['item_id:token']]
    total = df1.shape[0]
    likes_count = df1[df1['rating:float'] > threshold].shape[0]
    dislikes_count = total - likes_count
    liking_rate = round(likes_count/total, 4)
    return {'likes_count':likes_count,'dislikes_count':dislikes_count,'liking_rate':liking_rate}
```

```python
def cal_arr(ls,num):
    arr = []
    for i in range(0, len(ls), num):
        a1 = ls[i:i+num]
        if len(a1)==num:
            arr.append(a1)
    return arr
def build_user_inter(row):
    rs = row['rating:float']
    p_len = len([x for x in rs if x > threshold])
    p_rs = rs[:p_len]
    n_rs = rs[p_len:]

    ids = row['item_id']
    p_id_groups = ids[:p_len]
    n_id_groups = ids[p_len:]

    # 6p,4n
    p_num,n_num = 6,4
    p_rs1 = cal_arr(p_rs,p_num)
    n_rs1 = cal_arr(n_rs,n_num)

    p_id_groups1 = cal_arr(p_id_groups,p_num)
    n_id_groups1 = cal_arr(n_id_groups,n_num)
    
    # 对齐两个列表
    min_num = min(len(p_rs1), len(n_rs1))
    p_rs = p_rs1[:min_num]
    n_rs = n_rs1[:min_num]

    p_id_groups = p_id_groups1[:min_num]
    n_id_groups = n_id_groups1[:min_num]

    inter = {'p_id_groups': p_id_groups, 'n_id_groups': n_id_groups, 'p_rs': p_rs, 'n_rs': n_rs}
    return inter

inter_data['user_inter'] = inter_data.apply(lambda row: build_user_inter(row), axis=1)
inter_data.iloc[0]
```

```python
tr_temp = {
    "instruction": "Given the user's preference and unpreference, identify whether the user will like the target item by answering \"Yes.\" or \"No.\".",
    "input": "User Preference: {user_preference}\nUser Unpreference: {user_preference}\nWhether the user will like the target item \"{target_item}\"?",
    "output": ""
}
```

```python
def build_tr_data(data):
    tr_list = []
    for index, s_row in data.iterrows():
        row = tr_temp.copy()
        row['output'] = s_row['output']
        instruction = s_row['instruction']
        # print(row)
        t_arr = instruction.split('\n')
        like_start = [index for index, value in enumerate(t_arr) if value == 'Likes List:'][0]
        dislike_start = [index for index, value in enumerate(t_arr) if value == 'Dislikes List:'][0]
        target_start = [index for index, value in enumerate(t_arr) if value == 'Target Item:'][0]
        like_list = []
        dislike_list = []
        for t in t_arr[like_start+1:dislike_start-1]:
            t = f"{t[t.index('item name:')+11:t.index(', address:')]} ({t[t.index('category: ')+10:len(t)-1]})"
            like_list.append(t)
        for t in t_arr[dislike_start+1:target_start-1]:
            t = f"{t[t.index('item name:')+11:t.index(', address:')]} ({t[t.index('category: ')+10:len(t)-1]})"
            dislike_list.append(t)
        t = t_arr[target_start+1]
        target_item = f"{t[t.index('item name:')+11:t.index(', address:')]} ({t[t.index('category: ')+10:t.index(', likes count')]})"
        like_str,dislike_str,target_strt = str(like_list).replace('[','').replace(']','').replace('\'','\"'),str(dislike_list).replace('[','').replace(']','').replace('\'','\"'),str(target_item).replace('[','').replace(']','').replace('\'','\"')
        row['input'] = row['input'].replace('{user_preference}',like_str).replace('{user_preference}',dislike_str).replace('{target_item}',target_strt)
        tr_list.append(row)
        # print(row)
    return tr_list    
```

```python
tr_train_data = build_tr_data(train_data)
tr_test_data = build_tr_data(test_data)

l = int(len(tr_train_data)*6/8)
tr_valid_data = tr_train_data[l:]
tr_train_data = tr_train_data[:l]
tr_train_data,tr_valid_data,tr_test_data
```
